### LOAD CORPUS

```python
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader

# Loads the documents in from my data/raw directory
loader = DirectoryLoader(
    path="data/raw/",          
    glob="*.pdf",
    loader_cls=PyPDFLoader
)

documents = loader.load()
```

### COPUS STATISTICS REPORT

```python
print("Number of Documents: ", len(documents))

doc_lengths = [len(doc.page_content.split()) for doc in documents]
avg_doc_length = sum(doc_lengths) / len(documents) 
print("Average Document Length: ", avg_doc_length)

vocab = set()
for doc in documents:
    vocab.update(doc.page_content.lower().split())
print("Vocab Size: ", len(vocab))
```
